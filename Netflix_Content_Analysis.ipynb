```python
import pandas as pd

# pd → pandas의 별명
# read_csv() → CSV 파일을 표 형태로 읽는 함수
# on_bad_lines='skip' → 깨진 행은 건너뜀
# engine='python' → 파이썬 방식으로 읽음
df = pd.read_csv('mymoviedb (1).csv', on_bad_lines='skip', engine='python')

# len() → 행 수 세는 함수
# head() → 상위 5행만 미리보기
print(f'총 행 수: {len(df)}')
print(df.head())
```

```python
# dropna() → 결측치(빈 값) 있는 행 제거
df_genre = df.dropna(subset=['Genre'])

# str.split(', ') → 쉼표로 장르 분리
# explode() → 한 행에 있던 여러 장르를 각각 별도 행으로 펼침
df_genre = df_genre.assign(
    Genre=df_genre['Genre'].str.split(', ')
).explode('Genre')

print(f'원본 행 수: {len(df)}')
print(f'장르 쪼갠 후 행 수: {len(df_genre)}')
# value_counts() → 각 항목이 몇 번 나왔는지 세는 함수
# head(10) → 상위 10개만 보여줌
print(df_genre['Genre'].value_counts().head(10))
```

```python
# astype(float) → 문자로 저장된 숫자를 실수형으로 변환
# Vote_Count 10 미만 제거 → 신뢰도 낮은 평점 제외
df_clean = df_genre[df_genre['Vote_Count'].astype(float) >= 10]

# Vote_Average 0점 제거 → 투표 안 된 작품 제외
df_clean = df_clean[df_clean['Vote_Average'].astype(float) > 0]

# pd.to_datetime() → 날짜 형식으로 변환
# .dt.year → 연도만 추출
df_clean['Year'] = pd.to_datetime(df_clean['Release_Date']).dt.year

# to_csv() → 데이터프레임을 CSV 파일로 저장
# index=False → 행 번호는 저장 안 함
df_clean.to_csv('mymoviedb_clean.csv', index=False)
print(f'저장 완료! 최종 행 수: {len(df_clean)}')
```
