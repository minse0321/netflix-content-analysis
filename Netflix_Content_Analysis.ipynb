```
import pandas as pd

df = pd.read_csv('mymoviedb (1).csv', on_bad_lines='skip', engine='python')
print(f'총 행 수: {len(df)}')
print(df.head())
```

```
# 장르 쪼개기
df_genre = df.dropna(subset=['Genre'])
df_genre = df_genre.assign(Genre=df_genre['Genre'].str.split(', ')).explode('Genre')

# 결과 확인
print(f'원본 행 수: {len(df)}')
print(f'장르 쪼갠 후 행 수: {len(df_genre)}')
print(df_genre['Genre'].value_counts().head(10))
```

```
# 전처리 (결측치, 이상치 제거)
df_clean = df_genre[df_genre['Vote_Count'].astype(float) >= 10]
df_clean = df_clean[df_clean['Vote_Average'].astype(float) > 0]

# 연도 컬럼 추가
df_clean['Year'] = pd.to_datetime(df_clean['Release_Date']).dt.year

# CSV로 저장
df_clean.to_csv('mymoviedb_clean.csv', index=False)
print('저장 완료!')
print(f'최종 행 수: {len(df_clean)}')
```
